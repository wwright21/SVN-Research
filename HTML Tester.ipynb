```python
m=folium.Map(location=[34.22047316869057, -83.82039463228998], 
             tiles='OpenStreetMap', zoom_start=10)
# folium.TileLayer('Stamen Terrain', name = 'Terrain').add_to(m)

# folium.TileLayer('Stamen Water Color', name = 'Water Colors').add_to(m)
# folium.TileLayer('cartodbpositron').add_to(m)
# folium.TileLayer('cartodbdark_matter').add_to(m)
folium.TileLayer(
        tiles = 'https://mt1.google.com/vt/lyrs=s&x={x}&y={y}&z={z}',
        attr = 'Google',
        name = 'Google Satellite',
        overlay = False,
        control = True
    ).add_to(m)

folium.TileLayer(
        tiles = 'https://mt1.google.com/vt/lyrs=m&x={x}&y={y}&z={z}',
        attr = 'Google',
        name = 'Google Maps',
        overlay = False,
        control = True
    ).add_to(m)

folium.TileLayer(
        tiles = 'https://mt1.google.com/vt/lyrs=y&x={x}&y={y}&z={z}',
        attr = 'Google',
        name = 'Google Hybrid',
        overlay = False,
        control = True
    ).add_to(m)

folium.TileLayer(
    'Stamen Toner', 
    name = 'Black & White',
    overlay = False,
    control = True
    ).add_to(m)

folium.TileLayer(
        tiles = 'https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
        attr = 'Esri',
        name = 'Esri Satellite',
        overlay = False,
        control = True
       ).add_to(m)


folium.GeoJson(data = gaines_dt["geometry"],
               style_function = lambda x:
                   {'fill': True, 
                    'color': '#EE4B2B',
                    'fillColor':'#EE4B2B',
                    'fillOpacity':0.2},
               name = 'Gainesville 20-min Drivetime',
               overlay = True,
              tooltip = 'Gainesville 20-min DT'
              ).add_to(m)

folium.GeoJson(data = bras_dt["geometry"], 
               style_function = lambda x:
                   {'fill': True, 
                    'color': '#00FFFF',
                    'fillColor':'#00FFFF',
                    'fillOpacity':0.2},
               name = 'Braselton 20-min Drivetime',
              overlay = True,
              tooltip = 'Braselton 20-min DT').add_to(m)

HeatMap(a, radius=14, name='Pediatric Patient Heatmap',
       min_opacity=0.5).add_to(m)

# for coord in a:
#     folium.Marker(location=[ coord[0], coord[1] ], 
#                   fill_color='#43d9de', 
#                   radius=2).add_to(m)

folium.LayerControl().add_to(m)


m.save('Pediatrics_test.html')

m
```

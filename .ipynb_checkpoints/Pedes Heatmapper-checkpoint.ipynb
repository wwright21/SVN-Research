```python
import gmaps
import pandas as pd
from api_keys import g_key
```

```python
gmaps.configure(api_key=g_key)
```

```python
df1 = pd.read_csv('Pediatrics_Geocoded.csv')
df1
```

```python
locations = df1[['Lat','Long']].astype(float)

gmaps.configure(api_key=g_key)

fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, dissipating=False, max_intensity=25,
                                 point_radius = 0.01, opacity=0.5)

fig.add_layer(heat_layer)

fig
```
